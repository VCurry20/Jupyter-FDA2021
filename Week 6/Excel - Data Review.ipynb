We don't want the full spreadsheet - we only want certain rows and colums.

We can do this manually.

However within Pandas we also have options to cut the data down.

> https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html


When we read in the excel we can tell Pandas what we want / or what to leave out.

In this case we want to skip rows.

So this is an optional input (parameter) we can add which will give us a certain set of data to work with.

We can give a list of items, an integer or a callable option - which is what we would like

Excel is 1 indexed - Pandas / Python is 0 indexed

However the list option just skips x amount of data / rows

```python
df = pd.read_excel('http://www2.cao.ie/points/CAOPointsCharts2020.xlsx', skiprows=10) # df - skip first ten rows
```

```python
df  # you will see that it has cut out the first 10 rows of data - compare to the above
```
